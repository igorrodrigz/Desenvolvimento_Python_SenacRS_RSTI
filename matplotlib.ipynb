O que é **Matplot**?

Matplot é ma biblioteca para criação de gráficos 2D e 3D feita inteiramente em python.

*Exemplo simples de gráfico*

````python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show
````

## Anatomia da plotagem

##**Plotagem **

Plotar um gráfico significa unir os dados dos eixos e "desenhar" toda a estrutura. A principal função d matplot é plot(dadosX, dadosY). Ela é responsável por criar todo o gráfico.

Os valores máximos do gráfico são automaticamente ajustados pelo Matplot, porém, podemos customizar praticamente toda a plotagem.



```
# plt.plot(ypoints, marker = 'o')
```



````python
def criarTabela():
  print("Marker\t\tDescrição\t\tMarker\t\tDescrição")
  print(80*"-")
  print("'o'\t\tCircle\t\t'p'\t\tPentagon")
  print(80*"-")
  print("'*'\t\tStar\t\t'H'\t\tHexagon")
  print(80*"-")
  print("'.'\t\tPoint\t\t'h'\t\tHexagon")
  print(80*"-")
  print("','\t\tPixel\t\t'h'\t\tTriangle Down")
  print(80*"-")
  print("'x'\t\tX\t\t'^'\t\tTriangle Up")
  print(80*"-")
  print("'X'\t\tX(filled)\t'<'\t\tTriangle Left")
  print(80*"-")
  print("'+'\t\tPlus\t\t'D'\t\tDiamond")
  print(80*"-")
  print("'P'\t\tPlus(filled)\t'd'\t\tDiamond(thin)")
  print(80*"-")


criarTabela()


plt.plot(ypoints, marker = 'o')
plt.show()
````

## Linhas

As linhas dos dados também podem ser customizadas através do atributo `linestyle` ou `ls`.
Para as cores das linhas podemos também usar valores hexadecimais.

ls pode ser definido como:


*   `'solid'`
*   `'dotted'`
*   `'dashed'`
*   `'dashdot'`
*   `'None'`






````python
plt.plot(ypoints, linestyle = 'dotted', color = 'red')
plt.show()
````

## Grossura das Linhas

As linhas dos dados também podem ser customizadas através do atributo linestyle ou ls.

`linewidth` deve alterar a grossura das linhas.



````python
plt.plot(ypoints, linewidth = 20.5)
plt.show()
````

## Plotando várias linhas

A plotagem de múltiplas linhas é bem sugestiva. Basta plotar sequências de dados distintas em funções plot(). Veja o exemplo a seguir:

````python
y1 = np.array([3, 8, 1, 10])
y2 = np.array( [6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)
plt.show()
````

## Propriedades dos títulos e labels

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict=font1)
plt.xlabel("Average Pulse", fontdict=font2)
plt.ylabel("Calorie Burnage", fontdict=font2)

plt.plot(x, y)
plt.show()
````

## Inserindo Grades

````python
plt.plot(x, y)

plt.grid()

plt.show()
````

## Gráfico Scatter

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17])
y = np.array([99,86,87,88,111,86])

plt.scatter(x, y)
plt.show()
````

## Cor baseada em valor

Imagine o gradiente ao lado. É possível delimitar que os valores inferiores sejam mais roxos enquanto que os valores altos mais amarelados.

Primeiro criamos um array com valores de 0 a 100 que assumirão espaços de cores.


```
# colors = np.array([0. 10. 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
```




````python
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
c = colors
plt.scatter(x, y, cmap= 'viridis')

plt.colorbar()
````

## Tamanho baseado em valor

O tamanho segue a mesma idéia da cor, precisamos primeiro de um array de mesma quantidade que nossos dados. Cada valor representa o tamanho do ponto em seus respectivos x e y.

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8])
y = np.array([99,86,87])
sizes = np.array([20,50,100])

plt.scatter(x, y, s=sizes)
plt.show()
````

# Plotando gráficos de barras

A função plot() é trocada pela bar(). Além disso precisamos definir o que será escrito no eixo da label.

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
````

## Plotando histogramas

Um histograma é uma espécie de gráfico de barras que demonstra uma distribuição de frequências.
A função hist() é responsável por plotar o histograma.

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
````

## Plotando gráficos pizza

Com a função pie() podemos plotar o gráfico em pizza. Por padrão nenhum label para as partes será criado. Além disso, o array dos dados a serem plotados precisa resultar em 100, ou seja, 100% do espaço do gráfico.

### Labels:
O atributo labels permite inserir na ordem do array as labels para cada área da pizza.


````python
y = np.array([35,25,25,15])
mylabels = ["Maçã", "Banana", "Laranja", "Uva"]

plt.pie(y, labels = mylabels)
plt.show()
````
